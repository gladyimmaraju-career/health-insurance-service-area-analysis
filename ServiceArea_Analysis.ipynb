```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ServiceArea.csv")
df.head()
```

```python
df = df[['BusinessYear', 'StateCode', 'ServiceAreaName', 'County']]
df = df.dropna()
df.head()
```

```python
state_counts = df.groupby('StateCode')['ServiceAreaName'].count().sort_values(ascending=False)

plt.figure(figsize=(12,6))
state_counts.plot(kind='bar')

plt.title("Service Areas by State")
plt.xlabel("State")
plt.ylabel("Number of Service Areas")
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("chart1_state.png")
plt.show()
```

```python
year_counts = df.groupby('BusinessYear')['ServiceAreaName'].count()

plt.figure()
year_counts.plot(kind='line', marker='o')

plt.title("Service Areas by Year")
plt.xlabel("Year")
plt.ylabel("Count")

plt.savefig("chart2_year.png")
plt.show()
```

```python
county_counts = df.groupby('County')['ServiceAreaName'].count().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
county_counts.plot(kind='bar')

plt.title("Top 10 Counties by Service Areas")
plt.xlabel("County")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("chart3_county.png")
plt.show()
```

## Observations

- Service areas vary across different states
- Some counties have higher number of service areas
- The distribution is not uniform across regions
- Certain regions dominate in coverage
